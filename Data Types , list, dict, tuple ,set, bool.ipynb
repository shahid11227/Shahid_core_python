# Tuple
- Collection which can store elements of different data types
- created using `()` and elements are separated by `,`
- Tuples are like lists but they are immutable

```python
a = (1,2,3)
type(a)
```

```python
a[2] = 4                               # tuples are immutable
```

```python
t = (1,2,3,4,5.5,6.5, 2,3, 'ABC', [1,2,3],(1,2,3), {1 : 2}, True)
t
```

# Sets
- A Set in Python is a collection of unique items (no duplicates) that is unordered and mutable (can be changed).

Key Features:
- Unordered: No index.
- No duplicates.
- Can store multiple data types.
- Mutable (you can add/remove items).


```python
# A set is an unordered collection of unique elements
s = {1,2,3,4,5}
type(s)
```

```python
s = set()
type(s)
```

```python
s = {1,2,3,'ABC',4 ,5 ,'XYZ', 1.5}
s
```

```python
s = {1,2,3,4,2,3,4,5,6,5}
s
```

```python
s[0]
```

```python
s1 = {1,2,3,4,5}
s1
```

```python
s1.add(6)
s1
```

```python
s1.add(4)
s1
```

```python
s1.remove(4)
s1
```

```python
names = ['John','Bob','Tom','Bob', 'John','Babulal','Rahul', 'Babulal']
unique_names = set(names)
unique_names
```
