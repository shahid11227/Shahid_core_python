 #### Create a BankAccount class
- Attributes: account_number, balance
- Methods:
  -  deposit(amount)
  - withdraw(amount) 
  - display_balance()
  - Also add logic to prevent withdrawing more than the balance.

```python
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Insufficient balance!")

    def display_balance(self):
        print(f"Account {self.account_number} Balance: {self.balance}")

# Test
acc = BankAccount("12345")
acc.deposit(1000)
acc.withdraw(500)
acc.display_balance()
```

#### Create a Book class
- Attributes: title, author, pages

- Method: is_long_book() → returns True if pages > 300

```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def is_long_book(self):
        return self.pages > 300

# Test
b = Book("Python 101", "Mike", 350)
print("Is long book:", b.is_long_book())
```

#### Create a Student class
- Attributes: name, marks (a list of numbers)

##### Method:

- average() → returns average of marks

- get_grade() → returns:

  - "A" if avg ≥ 90

  - "B" if avg ≥ 75

  - "C" if avg ≥ 60

  - "F" otherwise  - 

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def average(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        avg = self.average()
        if avg >= 90:
            return "A"
        elif avg >= 75:
            return "B"
        elif avg >= 60:
            return "C"
        else:
            return "F"

# Test
s = Student("Sara", [85, 78, 92])
print("Average:", s.average())
print("Grade:", s.get_grade())
```
